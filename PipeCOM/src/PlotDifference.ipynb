# Import Libraries

```python
import os
import csv
import matplotlib.pyplot as plt
```

# Get CSV

```python
filename = "com_data.csv"

filepath = "../" + filename
```

# Print Lines

```python
csv_file = open(filepath, mode='r')
csv_reader = csv.reader(csv_file, delimiter=",")

for row in csv_reader:
    print(f"X: {row[0]}\nY: {row[1]}\nZ: {row[2]}\n")

csv_file.close()
```

# Show Graph

```python
x_difference = list()
y_difference = list()
z_difference = list()
frame_number = list()

csv_file = open(filepath, mode='r')
csv_reader = csv.reader(csv_file, delimiter=",")

next(csv_reader)

for row_num, row in enumerate(csv_reader):
    x_difference.append(float(row[0]))
    y_difference.append(float(row[1]))
    z_difference.append(float(row[2]))
    frame_number.append(int(row_num))

# Initialise the subplot function using number of rows and columns 
figure, axis = plt.subplots(3, 1, figsize=(6, 15)) 

axis[0].plot(frame_number, x_difference) 
axis[0].set_title("X Difference") 
axis[0].set_xlabel("Frame [No.]")
axis[0].set_ylabel("X Difference [mm]")

axis[1].plot(frame_number, y_difference) 
axis[1].set_title("Y Difference") 
axis[1].set_xlabel("Frame [No.]")
axis[1].set_ylabel("Y Difference [mm]")

axis[2].plot(frame_number, z_difference) 
axis[2].set_title("Z Difference") 
axis[2].set_xlabel("Frame [No.]")
axis[2].set_ylabel("Z Difference [mm]")

# Add padding  
figure.tight_layout(pad=1)

# Combine all the operations and display 
plt.show() 

csv_file.close()
```

## Show X Difference

```python
fig, ax = plt.subplots()
ax.scatter(frame_number, x_difference)
plt.show()
```
